## Q1. Make a calculator using Python with:
Addition
Subtraction
Multiplication
Division
Power

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("Power:", num1 ** num2)
```

## Q2. Write a program to check if there is any numeric value in a list using a for loop.

```python
my_list = [1, "hello", 3.5, "world"]

for item in my_list:
    if type(item) == int or type(item) == float:
        print("Numeric value found:", item)
```

## Q4. Write a Python program to sum all the numeric items in a dictionary.

```python
my_dict = {"a": 10, "b": 20, "c": 30}

total = 0

for value in my_dict.values():
    total = total + value

print("Sum is:", total)
```

## Q5. Write a program to identify duplicate values from a list.

```python
my_list = [1, 2, 3, 2, 4, 5, 1]

duplicates = []

for item in my_list:
    if my_list.count(item) > 1:
        if item not in duplicates:
            duplicates.append(item)

print("Duplicate values:", duplicates)
```

## Q6. Write a Python script to check if a given key already exists in a dictionary.

```python
my_dict = {"name": "Sara", "age": 22}

key = input("Enter key to check: ")

if key in my_dict:
    print("Key exists in dictionary.")
else:
    print("Key does not exist.")
```
