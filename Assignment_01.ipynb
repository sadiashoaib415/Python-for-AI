# 7. Write a program which takes 5 inputs from user for different subject's marks, total it and generate mark sheet using grades.

```python
# Taking 5 inputs
maths = int(input("Math: "))
eng = int(input("English: "))
sci = int(input("Science: "))
urdu = int(input("Urdu: "))
isl = int(input("Islamiat: "))

total = maths + eng + sci + urdu + isl
percent = (total / 500) * 100

if percent >= 80:
    grade = "A+"
elif percent >= 70:
    grade = "A"
elif percent >= 60:
    grade = "B"
else:
    grade = "Fail"

print(f"Total Marks: {total}/500 | Percentage: {percent}% | Grade: {grade}")
```

```python
english = int(input("Enter English Marks: "))
islamiat = int(input("Enter Islamiat Marks: "))
maths = int(input("Enter Maths Marks: "))
urdu = int(input("Enter Urdu Marks: "))
physics = int(input("Enter Physics Marks: "))

total = english + islamiat + maths + urdu + physics
percentage = (total / 500) * 100

if percentage >= 80:
    grade = "A+"
elif percentage >= 70:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 50:
    grade = "C"
else:
    grade = "Fail"

print("\n--- Mark Sheet ---")
print(f"Total Marks: {total}/500")
print(f"Percentage: {percentage}%")
print(f"Grade: {grade}")
```

# 8. Write a program which take input from user and identify that the given number is even or odd.

```python
num = int(input("Enter a number: "))
if (num % 2) == 0:
   print(f"{num} is Even")
else:
   print(f"{num} is Odd")
```
